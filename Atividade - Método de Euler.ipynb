### Item 1 - Resolvendo uma equação diferencial pelo método de Euler

Vamos continuar a atividade que você iniciou no papel. Você estava tentando encontrar pares de pontos que satisfizessem a seguinte equação diferencial:

$$ \frac{dY(t)}{dt}=1 \cdot Y(t)$$

Para resolver essa equação, que representa um modelo geométrico com taxa de crescimento de 100%, precisamos aproximá-la por uma equação a diferenças de passo $\Delta t$; quanto menor for o $\Delta t$, mais nos aproximaremos da solução da equação diferencial acima. A equação a seguir é a **aproximação** da equação diferencial:

$$\frac{Y(t+\Delta t)-Y(t)}{\Delta t}=1\cdot Y(t)$$

Vamos resolver essa equação diferencial no Python, aplicando o método de Euler. Para isso, vamos isolar o termo $Y(t+\Delta t)$:

$$Y(t+\Delta t)=Y(t)+1\cdot Y(t)\cdot \Delta t$$

A partir dessa equação, fazemos um loop e geramos uma série temporal a partir de $Y(0)=1$, obtendo uma **solução numérica** que é muito parecida com a **solução analítica** $Y(t)=Y(0)\cdot e^{t}$.

- $T_{max} = 3\,s$ (queremos analisar uma série que vai até 3s)
- $Y(0)=1$ (esse é o valor inicial de $Y$)

Agora é com você! Escreva seu código para gerar a lista de $Y(t)$, assim como você fez para gerar as populações.

### Item 2 - Resolvendo outra equação diferencial pelo método de Euler

Vamos agora utilizar o método de Euler para resolver outra equação diferencial. Note que, dessa vez, a expressão que define a derivada depende não só de $x$, como também do tempo $t$.

$$\frac{dx}{dt}= x^2 \cdot t - 2$$

A unidade de tempo utilizada é a **hora**.

#### a)
Implemente na célula abaixo a solução numérica desse sistema usando o método de Euler e plote em um gráfico a solução $x(t)$. Para isso, utilize uma lista de tempo de $0$ a $8\,h$ com $\Delta t=10^{-3}h$.

O valor inicial de $x(t)$ deve ser:

- $x(0)=-1$

```python
# SEU CÓDIGO DO ITEM 2A
# Bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt

# Lista de tempo
dt = 1e-3
Tmax = 8
tempo = np.arange(0,Tmax, dt)

# Lista de valores
x = [0]*len(tempo)

# Condições iniciais
x0 = -1
x[0] = x0

# Aplicando o método de Euler com o laço de repetição while
i = 0
while(i<len(tempo)-1):
    x[i+1] = x[i] + (x[i]**2 * tempo[i] - 2)*dt
    i = i+1
    
# Plotando o gráfico de x por t
plt.plot(tempo,x)
plt.xlabel("Tempo (h)")
plt.ylabel("x(t)")
plt.grid(True)
plt.show()

```
