### Item 1 - Implementando a sequência de Fibonacci em Python

A famosa sequência de Fibonacci é caracterizada da seguinte forma: nas duas primeiras posições, aparece o número 1. Da terceira posição em diante, cada termo é dado pela soma dos dois anteriores. No código a seguir, veja que foi criada uma lista denominada `fib`, de 10 posições, com os 2 primeiros termos já definidos.

Preencha os valores `fib[2]`, `fib[3]` e assim por diante. Porém, você deve fazer isso **sem digitar diretamente os valores da sequência**, ou seja, escrevendo para cada termo uma linha de código que o calcule em função dos termos anteriores.

Ao final de sua implementação, imprima a lista `fib`.

```python
fib=[2,7,9]
print(fib)

fib = [1]*10
print(fib)
fib[0] = 1
fib[1] = 1
fib[2] = fib[0] + fib[1]
fib[3] = fib[1] + fib[2]
fib[4] = fib[2] + fib[3]
fib[5] = fib[3] + fib[4]
fib[6] = fib[4] + fib[5]
fib[7] = fib[5] + fib[6]
fib[8] =fib[6] + fib[7]
fib[9] = fib[7] + fib[8]
print(fib)
```

Você deve ter pensado: "Ainda bem que foram só os 10 primeiros termos da sequência de Fibonacci. Imagine fazer isso para 100 termos!". Pois é, é o que vamos pedir que você faça na sequência! Calma, você não fará isso na "mão".

### Item 2 - Implementando seu primeiro loop em Python

Linguagens de programação têm estruturas lógicas de repetição, denominadas "loops". Em ModSim, vamos usar duas delas:

- `while`
- `for`

Vamos começar pela estrutura `while`. Veja a explicação do seu professor antes de fazer o que é pedido a seguir.

Gere, na célula abaixo, um código que preencha os 100 primeiros termos da sequência de Fibonacci. Imprima a lista de Fibonacci e perceba como os valores crescem de forma acelerada.

```python
fib=[0]*100
fib[0] = 1
fib[1] = 1
i=0
while (i <= 97):
    fib[i+2]=fib[i+1]+fib[i]
    i=i+1

print(fib)
```

### Item 3 - Implementando seu primeiro gráfico em Python

Python é uma linguagem de alto nível dedicada à ciência e à engenharia, por isso já temos à nossa disposição módulos gráficos e matemáticos, assim como algoritmos de vários tipos e funcionalidades. Incrível, não?
Um dos módulos que vamos usar sempre é o `pyplot` da biblioteca `matplotlib`, que é uma biblioteca para a geração de gráficos.

Com o intuito de entender como usar esse módulo, vamos gerar o gráfico da função de lei

$$f(x)=x^2$$

com $0 \le x \le 2$.

Para fazer isso em Python, vamos precisar de duas listas: uma com os valores de $x$ e outra com os correspondentes valores de $y$, que são dados por $y=x^2$.

Uma forma de gerar a lista com os valores de $x$ é utilizar a função `arange` da biblioteca `numpy`. Primeiramente, precisamos importar essa biblioteca escrevendo:

`import numpy as np`

Essa forma de usar o `import` permite agilizar a escrita do código: em vez de escrevermos `numpy.arange` todas as vezes que precisarmos desse comando, basta escrever `np.arange` (ou seja, definimos que `np` é o "apelido" da biblioteca `numpy`).

#### a)
Observe e rode o código da célula abaixo para entender a sintaxe da função `arange`.

```python
import numpy as np

lista_x = np.arange(0,2.1,0.1)

print(lista_x)
```

#### b)
Agora, você vai gerar a lista com os valores de $y$ associados a cada um dos valores de $x$ da `lista_x` que acabamos de definir. Para isso, será necessário usar o loop `while`.

Comece declarando uma lista de zeros chamada `lista_y` com o mesmo tamanho da `lista_x` e escreva um loop que prencha essa lista com os valores de $y=x^2$. Em seguida, imprima a `lista_y` e verifique se os valores condizem com o esperado.

```python
lista_x=np.arange(0,19,1)
lista_y=[0]*len(lista_x)
i=0
while(i<len(lista_x)):
    lista_y[i]=int(lista_x[i]**2)
    i=i+1
print(lista_y)
```

#### c)
Por fim, plote o gráfico de $y=f(x)$ em função de $x$. Para isso, você vai precisar importar o módulo `pyplot`:

`import matplotlib.pyplot as plt`

Em seguida, basta usar o comando `plt.plot( lista_x , lista_y )`. Com `plt.xlabel("Título do eixo x")` e `plt.ylabel("Título do eixo y")` damos títulos aos eixos; para exibir a malha quadriculada, usamos `plt.grid()`

```python
import matplotlib.pyplot as plt
plt.plot( lista_x , lista_y )
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
```

### Item 4 - Implementando um modelo populacional aritmético

Agora, vamos aplicar os conhecimentos adquiridos nos itens anteriores nos três principais modelos de crescimento populacional que vimos em aula.

A figura abaixo mostra um modelo aritmético.



Como vimos, a equação a diferenças desse modelo é:

$P(t+1)=P(t)+N-M$

Vamos considerar os seguintes parâmetros para o nosso modelo:

$P(0)=15 \space$ [onças]

$N=4 \space$ [onças/mês]

$M=1 \space$ [onças/mês]



#### a)

Usando a equação a diferenças do modelo, construa a série temporal da população de onças para 40 meses.

```python
P=list(range(40))
print (P)
P[0]=15
i=0
while(i<39):
    P[i+1]=P[i]+4-1
    i=i+1

print(P)
```

#### b) 

Em seguida, plote um gráfico do número de onças em função do tempo.

```python
import numpy as np
tempo=np.arange(0,41,1)
import matplotlib.pyplot as plt
plt.plot( tempo , P , "ro")
plt.xlabel("Tempo")
plt.ylabel("População")
plt.grid()
```

### Item 5 - Implementando um modelo populacional geométrico

Considere agora o diagrama de um modelo populacional geométrico, conforme figura abaixo.



Você já deve estar bem familiarizado(a) com  a equação a diferenças desse modelo, que é:

$P(t+1)=P(t)+p_n \cdot P(t)-p_m\cdot P(t)$

Vamos considerar os seguintes parâmetros para o nosso modelo:

$P(0)=25 \space$ [onças]

$p_n=0,1 \space$ [onças/onça/mês]

$p_m=0,05 \space$ [onças/onça/mês]

#### a)

Usando a equação a diferenças do modelo, construa a série temporal da população de onças para 40 meses.

```python
P=list(range(41))
P[0]=25
i=0
while(i<40):
    P[i+1]=P[i]*1.05
    i=i+1

print(P)
```

#### b)

Plote um gráfico do número de onças em função do tempo.

```python
import numpy as np
tempo=np.arange(0,41,1)
print(tempo)
import matplotlib.pyplot as plt
plt.plot( tempo , P , "ro")
plt.xlabel("Tempo")
plt.ylabel("População")
plt.grid()

```

### Item 6 - Implementando um modelo populacional logístico

Se você chegou até aqui, talvez queira ir mais longe!
Considere agora o diagrama de um modelo logístico, conforme figura abaixo.



Trabalhamos esse modelo na última aula, e você deve se lembrar de que sua equação a diferenças é:

$P(t+1)=P(t)+\alpha \cdot P(t)-\alpha\cdot \dfrac{P(t)^2}{S}$

Vamos considerar os seguintes parâmetros para o modelo logístico:

$P(0)=5 \space$ [onças]

$\alpha=0,2 \space$ [onças/onça/mês]

$S=200 \space$ [onças]



####a)

Usando a equação a diferenças do modelo, construa a série temporal de onças para 40 meses.

```python
P=list(range(41))
P[0]=5
i=0
while(i<40):
    P[i+1]=P[i] + P[i]*0.2-0.2*P[i]**2/200
    i=i+1

print(P)

```

#### b)

Plote um gráfico do número de onças em função do tempo.

```python
import numpy as np
tempo=np.arange(0,41,1)
print(tempo)
import matplotlib.pyplot as plt
plt.plot( tempo , P , "ro")
plt.xlabel("Tempo")
plt.ylabel("População")
plt.grid()

```
